## TEXT ANNOTATIONS

* The uses of the basic text() will place text at an arbitrary position on the Axes.
* A common use case of text is to annotate some feature of the plot, and the annotate() method provides helper functionality to make annotations easy. 
* In an annotation, there are two points to consider: the location being annotated represented by the argument xy and the location of the text xytext. Both of these arguments are (x,y) tuples.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
population_in_village=[534,689,258,401,724,689,435,534,545,435,423,789,905,561]
ids=range(1,15)

y_coordinate=min(population_in_village)
x_coordinate=ids[population_in_village.index(y_coordinate)]

plt.plot(ids,population_in_village)
plt.xlabel("ids")
plt.ylabel("population in numbers")
plt.title("Text annotation example")
plt.annotate('lowest value',xy=(x_coordinate,y_coordinate),xytext=(5,300),arrowprops=dict(facecolor='black',shrink=0.05))
plt.show()                                                                                  #lower the shrink,bigger is the arrow size
```
