## ADD TITLE AND AXIS LABELS

syntax:
* xlabel("string") - It sets the label for x-axis which is specified by the string
* ylabel("string") - It sets the label for y-axis which is specified by the string
* title("string") - This will set the title of the plot as specified by the string

```python
#using title,xlabel,ylabel
import matplotlib.pyplot as plt
%matplotlib inline

x=[1,2,3]
y=[1,4,9]
plt.plot(x,y)
plt.title("our first plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
