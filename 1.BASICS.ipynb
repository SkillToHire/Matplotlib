# MATPLOTLIB

* Matplotlib is an open source python visualization library
* This is a powerful plotting library useful for those working with Python and Numpy
* The module used in Matplotlib is pyplot

```python
#import matplotlib library
import matplotlib.pyplot as plt
```

## BASICS

### Simple plot

syntax:
* plot(x-axis values,y-axis values) - This will plots a simple line graph with x-axis values against y-axis values
* show() -This will displays the graph

```python
x=[1,2,7,9,10]
plt.plot(x)
plt.show()
```

### Plot with x and y values

```python
#simple plot

x=[1,2,3]
y=[1,4,9]
plt.plot(x,y)
plt.show()
```

```python
#other way to plot

plt.plot([1,2,3],[1,4,9])
plt.show()
```
