## SETTING UP GRIDS IN OUR PLOTS

* syntax:
    The grid() function of axes object sets visibility of grid inside the figure to on or off. 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r',marker='o') 
plt.plot(x,z,label='z',color='k',marker='s')
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#using linewidths for grids

x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r',marker='o') 
plt.plot(x,z,label='z',color='k',marker='s')
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.grid(True,linewidth=5)
plt.show()
```

```python
#we can add color to the grid

x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r',marker='o') 
plt.plot(x,z,label='z',color='k',marker='s')
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.grid(True,linewidth=1,color='c')
plt.show()
```

```python
#we can add linestyle to the grid

x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r',marker='o') 
plt.plot(x,z,label='z',color='k',marker='s')
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.grid(True,linewidth=1,color='c',linestyle='-.')
plt.show()
```
