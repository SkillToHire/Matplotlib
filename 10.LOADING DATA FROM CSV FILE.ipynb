## LOADING DATA FROM CSV FILE

```python
#loading data 
import pandas as pd

data=pd.read_csv("sample_data.csv")
data
```

```python
type(data) #This is a dataframe type defined by pandas module
```

```python
type(data.column_a) #This is a series type defined by pandas module
```

```python
#plot
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(data.column_c,data.column_b)
plt.show()
```

```python
#using xlabel,ylabel,title
plt.plot(data.column_c,data.column_b)
plt.xlabel("x")
plt.ylabel("y")
plt.title("simple plot")
plt.show()
```

```python
#to draw multiple lines on same plot

plt.plot(data.column_b,data.column_a)
plt.plot(data.column_b,data.column_c)
plt.xlabel("x")
plt.ylabel("y")
plt.title("simple plot")
plt.show()
```

```python
#using legend

plt.plot(data.column_b,data.column_a)
plt.plot(data.column_b,data.column_c)
plt.xlabel("x")
plt.ylabel("y")
plt.title("simple plot")
plt.legend(["b vs a","b vs c"])
plt.show()
```
