## SETTING X AND Y AXES LIMITS

* syntax:
    * xlim() =>Get or set the x limits of the current axes.
    * ylim() =>Get or set the y limits of the current axes.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#we set x and y axes limits using xlim and ylim

x=[30,10,20]
y=[40,50,60]
z=[30,40,50]
plt.plot(x,y,label='y',color='r',marker='o',markerfacecolor='b') 
plt.plot(x,z,label='z',color='k',marker='s',markerfacecolor='g')
plt.xlim(10,50)
plt.ylim(10,50)
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.show()
```
