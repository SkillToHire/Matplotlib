## ADD LEGEND FOR PLOTS

syntax:
* legend(loc) â€” This is used to make legend of the graph #default=upper left
* we can change the location using loc

```python
#using legend will make us able to differentiate between two lines by using different color
import matplotlib.pyplot as plt
%matplotlib inline

x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y)
plt.plot(x,z)
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["y","z"])
plt.show()
```

```python
#other way to mention labels of legend

x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y')
plt.plot(x,z,label='z')
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.show()
```
