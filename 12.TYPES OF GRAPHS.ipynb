## TYPES OF GRAPHS

### BARPLOT

* Barplot is a frequency plot for qualitative variable(or categorical variable).
* Barplot can be used to assess the most-occurring and least-occurring categories within a 
  data set.
* syntax:
bar(categorical variables, values, color) â€” It is used to create vertical bar graphs

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=[1,5,8,9]
y=[5,3,8,7]
plt.bar(x,y)
plt.title("Barplot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#we can change the color

x=[1,5,8,9]
y=[5,3,8,7]
plt.bar(x,y,color='red')
plt.title("Barplot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#multiple bars on same plot

x1=[2,4,6,8]
y1=[6,7,8,2]

x2=[1,3,5,7]
y2=[7,8,2,4]
plt.bar(x1,y1)
plt.bar(x2,y2)

plt.title("Barplot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["bar1","bar2"])
plt.show()
```

```python
#example
temp_city=[20,30,21,24,25,34,45,33,21,22,50,45,46,35,47,33,34,35,43,46,21,42,53,54,44,39,30,40,28,34,21,22,25,43,42,43]
ids=[i for i in range(len(temp_city))]

plt.bar(ids,temp_city)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#bar graph of sample data
import pandas as pd
data=pd.read_csv("sample_data.csv")
print(data)

plt.bar(data["column_a"],data["column_b"],color='r')
plt.bar(data["column_a"],data["column_c"],color='b')
plt.show()
```

### HISTOGRAM

* Histogram is the visual representation of the data which can be used to assess the probability distribution(frequency distribution) of the data.
* syntax:
hist(values, number of bins) â€” It is used to create a histogram

```python
x=[1,4,6,7,8]
y=[10,4,7,3,2]
plt.hist(y)
#plt.hist(x)
plt.show()
```

```python
#example
temp_city=[20,30,21,24,25,34,45,33,21,22,50,45,46,35,47,33,34,35,43,46,21,42,53,54,44,39,30,40,28,34,21,22,25,43,42,43]

plt.hist(temp_city)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x=np.random.randn(1000)
plt.title("Histogram")
plt.xlabel("Random Data")
plt.ylabel("Frequency")
plt.hist(x,10)
plt.show()
```

```python
#histogram of sample data

data=pd.read_csv("sample_data.csv")
print(data)

plt.hist(data["column_a"],data["column_b"],color='r')
plt.show()
```

### PIE CHARTS

* Pie chart is mainly used for categorical data and is a circular chart that displays the proportion of each category in the data set.
* syntax:
pie(value, categorical variables) â€”It is used to create a pie chart

```python
#simple pie chart

labels=["dogs","cats","horses","fishes"]
sizes=[60,45,20,10]

plt.pie(sizes,labels=labels)
plt.show()
```

```python
#example
firms=["Firm_A","Firm_B","Firm_C","Firm_D","Firm_E"]
market_share=[20,25,15,10,20]
Explode=[0,0.1,0,0,0] ##only "explode" the 2nd slice (i.e. 'Firm_B')
plt.pie(market_share,explode=Explode,labels=firms,shadow=True,startangle=40)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.legend(title="List of Firms")
plt.show()
```

### SCATTER PLOTS

* Scatter plots are used to plot data points on horizontal and vertical axis in the attempt to show how much one variable is affected by another.
* syntax:
scatter(x-axis values, y-axis values) â€” It plots a scatter plot with x-axis values against y-axis values

```python
#simple scatter plot

x=[1,2,3,4,5,6,7,8]
y=[2,4,5,9,3,10,8,1]

plt.scatter(x,y)
plt.title("scatter plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8]
y=[2,4,5,9,3,10,8,1]

plt.scatter(x,y,marker="*",s=500) #s=size of the marker of * 
plt.title("scatter plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

* Here we feed in arbitrarily created data of height and weight and plot them against each other.
* We used xlim() and ylim() methods to set the limits of X-axis and Y-axis respectively.

```python
#example
height = np.array([167,170,149,165,155,180,166,146,159,185,145,168,172,181,169])
weight = np.array([86,74,66,78,68,79,90,73,70,88,66,84,67,84,77])
plt.xlim(140,200)
plt.ylim(60,100)
plt.scatter(height,weight)
plt.title("scatter plot")
plt.xlabel("Height")
plt.ylabel("weight")
plt.show()
```

### STACK PLOTS

* Stack plot visualizes data in stacks and shows the distribution of data over time.
* It is used for checking multiple variable area plots in a single plot.

```python
## stack plots

a=[1,2,3,4,5]
b=[1,2,5,6,7]
c=[3,2,5,4,6]
d=[7,8,5,6,7]

e=np.vstack([b,c,d])
plt.stackplot(a,e,colors=["m","c","r"])

plt.title("stack plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
