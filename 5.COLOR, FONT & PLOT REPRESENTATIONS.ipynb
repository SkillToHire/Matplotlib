## COLOR,FONTS AND PLOT REPRESENTATION

* colors:-
  * b: blue
  * g: green
  * r: red
  * c: cyan
  * m: magenta
  * y: yellow
  * k: black
  * w: white

* markers:-
  * "." =>point
  * "," =>pixel
  * "o" =>circle
  * "v" =>triangle_down
  * "^" =>triangle_up
  * "<" =>triangle_left
  * ">" =>triangle_right
  * "1" =>tri_down
  * "2" =>tri_up
  * "3" =>tri_left
  * "4" =>tri_right
  * "8" =>octagon
  * "s" =>square
  * "p" =>pentagon
  * "P" =>plus (filled)
  * "*" =>star
  * "h" =>hexagon1
  * "H" =>hexagon2
  * "+" =>plus
  * "x" =>x
  * "X" =>x (filled)
  * "D" =>diamond
  * "d" =>thin_diamond
  * "|" =>vline
  * "_" =>hline
  * 0 (TICKLEFT) =>tickleft
  * 1 (TICKRIGHT) =>tickright
  * 2 (TICKUP) =>tickup
  * 3 (TICKDOWN) =>tickdown
  * 4 (CARETLEFT) =>caretleft
  * 5 (CARETRIGHT) =>caretright
  * 6 (CARETUP) =>caretup
  * 7 (CARETDOWN) =>caretdown
  * 8 (CARETLEFTBASE) =>caretleft (centered at base)
  * 9 (CARETRIGHTBASE) =>caretright (centered at base)
  * 10 (CARETUPBASE) =>caretup (centered at base)

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#using color
x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r')
plt.plot(x,z,label='z',color='k')
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.show()
```

```python
#marker
x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r',marker='o') #circle marker
plt.plot(x,z,label='z',color='k',marker='s')#square marker
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.show()
```

```python
#we can change the color of the marker using markerfacecolor

x=[1,2,3]
y=[1,4,9]
z=[4,5,6]
plt.plot(x,y,label='y',color='r',marker='o',markerfacecolor='b') #circle marker
plt.plot(x,z,label='z',color='k',marker='s',markerfacecolor='g')#square marker
plt.title("second plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend()
plt.show()
```
